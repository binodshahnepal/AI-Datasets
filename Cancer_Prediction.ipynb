### Removal of the features that have high correlation
Step 1: We calculate the correlation matrix, which gives the correlation coefficients between all pairs of features. Using abs() ensures we consider the magnitude of the correlation, regardless of direction (positive or negative).

Step 2: We isolate the upper triangle of the correlation matrix using np.triu() because the correlation matrix is symmetric, and we don't need to check both triangles.

Step 3: We loop through the columns of the upper triangle and identify features with a correlation greater than 0.9.

Step 4: We drop the highly correlated features, keeping just one from each pair.

```python
corr_matrix = df.corr().abs()  # Get absolute value of correlations
# Step 2: Identify pairs of highly correlated features (correlation > 0.9)
upper_tri = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)
```

```python
to_drop = [
    column for column in upper_tri.columns if any(upper_tri[column] > 0.9)
]

# Step 4: Drop the features
df_reduced = df.drop(columns=to_drop)
print(f"Features removed: {to_drop}")
print(f"Remaining features: {df_reduced.columns}")
```

```python
result={'Model Name':[],'Accuracy':[],'Recall':[],'F1 Score':[],'Precision':[]}
def evaluate_models(models, X_train, y_train, X_test, y_test):
    num_models = len(models)
    num_cols = 3
    num_rows = (num_models + num_cols - 1) // num_cols  # Compute rows needed for subplots

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 5, num_rows * 5))
    axes = axes.flatten()  # Flatten the 2D array of axes to easily iterate over

    for i, (name, model) in enumerate(models):
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        cm = confusion_matrix(y_test, y_pred)
        accuracy = accuracy_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')
        precision = precision_score(y_test, y_pred, average='macro', zero_division=1)
        result['Model Name'].append(name)
        result['Accuracy'].append(accuracy)
        result['Recall'].append(recall)
        result['Precision'].append(precision)
        result['F1 Score'].append(f1)
        # Plot confusion matrix
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
        disp.plot(ax=axes[i], cmap=plt.cm.Blues, colorbar=True)
        axes[i].set_title(name)

    # Remove unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
evaluate_models(models, X_train, y_train, X_test, y_test)
```
